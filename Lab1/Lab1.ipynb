Набор данных содержит 104 колонки, где отмечены различные характеристики каждого футболиста

```python
df = pd.read_csv('players_20.csv')
```

```python
df.head()
```

## График №1 - Рейтинг футболиста относительно возраста

```python
sns.set(rc = {"figure.figsize":(8,7)})
```

```python
sns.set_style("darkgrid")
plt.xlabel('Age')
plt.ylabel('Overall')
sns.lineplot(data=df, x="age", y="overall")
```

## График №2 - Топ-10 стран по количеству игроков

```python
data = df.groupby('nationality').agg('count')['player_url'].reset_index().sort_values('player_url', ascending=False)
```

```python
data
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=data.head(10), x='nationality', y='player_url')
plt.xlabel('Nationality')
plt.ylabel('Number of players')
plt.title('Top 10 countries by number of players', fontdict={'fontsize': 16})
plt.show()
```

## График №3 - Соотношение роста и веса футболистов

```python
sns.relplot(data=df, x='height_cm', y='weight_kg', height=10)
plt.xlabel('Height, cm')
plt.ylabel('Weight, kg')
plt.title('The ratio of height and weight of football players', fontdict={'fontsize': 16})
plt.show()
```

## График №4 - Матрица корреляций

```python
df.head(1)
```

```python
sns.heatmap(data = df.corr(numeric_only=True), cmap='mako')
```

## График №5 - Топ-20 футболистов с наивысшим рейтингом

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=df.head(20), x='short_name', y='overall')
plt.xlabel('Football player')
plt.ylabel('Rating')
plt.xticks(rotation=60)
plt.title('Top-20 highly rated football player', fontdict={'fontsize': 16})
plt.show()
```

## График №6 - Топ-10 футбольных клубов (по среднему рейтингу футболистов в нем)

```python
data = df.groupby('club').agg('mean', numeric_only=True)['overall'].reset_index().sort_values('overall', ascending=False)
```

```python
data.head()
```

```python
plt.figure(figsize=(15,10))
sns.pointplot(data=data.head(10), x='club', y='overall')
plt.xlabel('Football club')
plt.xticks(rotation='45')
plt.ylabel('Average rating')
plt.title('Top 10 football clubs', fontdict={'fontsize': 16})
plt.show()
```
