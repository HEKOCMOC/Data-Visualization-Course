### Что делает ```geom_col()```? Чем он отличается от ```geom_bar()```?

### Ответ:
```geom_col()``` представляет значения данных в виде столбцов. <br>
```geom_bar()``` делает высоту полосы пропорциональной количеству наблюдений в каждой группе. 

````python
ggplot(data=diamonds) +\
geom_col(mapping=aes(x=diamonds["cut"], y=diamonds["depth"]))
````

### Что произойдёт, если 
```ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", y="..prop.."), group=1)```

или
```ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", fill="color", y="..prop.."))```
### запустить без ```group=1```? Почему?

````python
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="clarity", fill="cut"), group=1)
````

````python
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="clarity", fill="cut"))
````

#### Ответ:
Параметр ```group=1``` приводит все типы к одной единой группе

### Зачем тут coord_fixed()? Что делает geom_abline()?

````python
ggplot(data=mpg, mapping=aes(x="cty", y="hwy")) +\
geom_point() +\
geom_abline() +\
coord_fixed()
````

#### Ответ:
```coord_fixed()``` делает фиксированным соотношение между масштабами x и y, а ```geom_abline()``` определяет уклон и пересечение линии.
